```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Define the labels for the axes (indices)
labels = ['Purchasing Power Index', 'Safety Index', 'Health Care Index', 'Cost of Living Index', 
          'Property Price to Income Ratio', 'Traffic Commute Time Index', 'Pollution Index', 'Climate Index']

# Number of variables
num_vars = len(labels)

# Simulated data for each continent (values between 0 and 1)
asia = [0.7, 0.6, 0.7, 0.4, 0.5, 0.7, 0.8, 0.6]
europe = [0.8, 0.9, 0.8, 0.6, 0.6, 0.5, 0.4, 0.7]
americas = [0.6, 0.7, 0.7, 0.5, 0.7, 0.6, 0.5, 0.6]
africa = [0.4, 0.5, 0.4, 0.3, 0.4, 0.4, 0.6, 0.7]
oceania = [0.7, 0.8, 0.9, 0.5, 0.6, 0.6, 0.3, 0.8]

# Create a list of these data arrays
data = [asia, europe, americas, africa, oceania]

# Compute angle of each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make the radar chart close to a circle (by adding the first value at the end of each data list)
asia += asia[:1]
europe += europe[:1]
americas += americas[:1]
africa += africa[:1]
oceania += oceania[:1]
angles += angles[:1]  # Add the first angle to close the circle

# Define different shades of blue for each continent
colors = ['#1f77b4', '#4682b4', '#5f9ea0', '#6495ed', '#87ceeb']  # Different shades of blue

# Plot the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Plot each continent's data with different shades of blue
ax.plot(angles, asia, color=colors[0], linewidth=2, label='Asia')
ax.fill(angles, asia, color=colors[0], alpha=0.25)

ax.plot(angles, europe, color=colors[1], linewidth=2, label='Europe')
ax.fill(angles, europe, color=colors[1], alpha=0.25)

ax.plot(angles, americas, color=colors[2], linewidth=2, label='Americas')
ax.fill(angles, americas, color=colors[2], alpha=0.25)

ax.plot(angles, africa, color=colors[3], linewidth=2, label='Africa')
ax.fill(angles, africa, color=colors[3], alpha=0.25)

ax.plot(angles, oceania, color=colors[4], linewidth=2, label='Oceania')
ax.fill(angles, oceania, color=colors[4], alpha=0.25)

# Set the labels for the axes
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])  # Remove the last angle since it's just for closing
ax.set_xticklabels(labels, ha='center')

# Add a title
plt.title('Radar Chart of Indexes by Continent', size=15, color='blue', weight='bold')

# Adjust the legend to prevent overlap and move it outside
plt.legend(loc='upper left', bbox_to_anchor=(1.1, 1), title='Continents')

# Display the chart
plt.tight_layout()  # Ensures everything fits without overlap
plt.show()

```
